Q1.For a sample space of 15 people, a statistician wanted to know the consumption of water and
other beverages.

 He collected their average consumption of water and beverages for 30 days (in
litres).

Help him to visualize the data using density plot, rug plot and identify the mean, median,
mode and skewness of the data from the plot.
image.png

```python
#Creating a dataframe
consumption = pd.DataFrame({"Water":[3.2, 3.5, 3.6, 2.5, 2.8, 5.9, 2.9, 3.9, 4.9, 6.9, 7.9, 8.0, 3.3, 6.6, 4.4],
"Beverages":[2.2, 2.5, 2.6, 1.5, 3.8, 1.9, 0.9, 3.9, 4.9, 6.9, 0.1, 8.0, 0.3, 2.6, 1.4]})

consumption.T #transpose of the dataframe
```

####**Density Plot**

```python
sns.distplot(consumption["Water"], hist=False,kde=True,
             bins=int(180/5), color = 'red',
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
mean_w = consumption["Water"].mean()
median_w = consumption["Water"].median()
mode_w = consumption["Water"][np.argmax(np.bincount(np.round(consumption["Water"]).astype(int)))]


plt.axvline(mean_w, color='red', linestyle='--', label=f'Mean: {mean_w:.2f}')
plt.axvline(median_w, color='blue', linestyle='-', label=f'Median: {median_w:.2f}')
plt.axvline(mode_w, color='green', linestyle='-', label=f'Mode: {median_w:.2f}')
plt.legend()
plt.show()
```

```python
sns.distplot(consumption["Beverages"], hist=False, kde=True,
             bins=int(180/5), color = 'green',
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
mean_b = consumption["Beverages"].mean()
median_b = consumption["Beverages"].median()
mode_b = consumption["Beverages"][np.argmax(np.bincount(np.round(consumption["Beverages"]).astype(int)))]

plt.axvline(mean_b, color='red', linestyle='--', label=f'Mean: {mean_b:.2f}')
plt.axvline(median_b, color='blue', linestyle='-', label=f'Median: {median_b:.2f}')
plt.axvline(mode_b, color='green', linestyle='-', label=f'Mode: {median_w:.2f}')
plt.legend()
plt.show()
```

```python
sns.distplot(consumption["Beverages"], hist=False, kde=True,
             bins=int(180/5), color = 'green',
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})

plt.axvline(mode_w, color='yellow', linestyle='dashdot', label=f'Skewness-w: {median_w:.2f}')
plt.axvline(mode_b, color='pink', linestyle='dashdot', label=f'Skewness-b: {median_b:.2f}')
plt.legend()
plt.show()
```

#### Calculating the Skewness

```python
skewness_w = np.round((3 * (mean_w - median_w)) / np.std(consumption["Water"]), 2)
skewness_b = np.round((3 * (mean_b - median_b)) / np.std(consumption["Beverages"]), 2)

print("Skewness for Water Column : ",skewness_w)
print("Skewness forBeverages Column : ",skewness_b)
# Add mean, median, mode, and skewness to the plots
```

```python
sns.kdeplot(consumption["Water"], label='Water', shade=True,color ="red")
sns.kdeplot(consumption["Beverages"], label='Beverages', shade=True,color="green")

mean_w = consumption["Water"].mean()
mean_b = consumption["Beverages"].mean()

print(mean_w,mean_b)

#plotting as a whole
plt.axvline(mean_w, color='red', linestyle='--', label=f'Mean-w: {mean_w:.2f}')
plt.axvline(mean_b, color='blue', linestyle='--', label=f'Mean-b: {mean_b:.2f}')
plt.axvline(median_w, color='red', linestyle='-', label=f'Median-w: {median_w:.2f}')
plt.axvline(median_b, color='blue', linestyle='-', label=f'Median-b: {median_b:.2f}')
plt.axvline(mode_w, color='red', linestyle=':', label=f'Mode-w: {median_w:.2f}')
plt.axvline(mode_b, color='blue', linestyle=':', label=f'Mode-b: {median_b:.2f}')
plt.axvline(mode_w, color='yellow', linestyle='dashdot', label=f'Skewness-w: {median_w:.2f}')
plt.axvline(mode_b, color='pink', linestyle='dashdot', label=f'Skewness-b: {median_b:.2f}')

plt.xlabel('X-axis Label')
plt.ylabel('Density')
plt.title('Density Plot of Water and Beverages')

plt.legend()


plt.show()
```

**Q2 A car company wants to predict how much fuel different cars will use based on their masses.**
**They took a sample of cars, drove each car 100km, and measured how much fuel was used in each case (in litres).**

**Visualize the data using scatterplot and also find co-relation between the 2
variables (eg. Positive//Negative, Linear/ Non-linear co-relation) The data is summarized in the table below.**


*(Use a reasonable scale on both axes and put the explanatory variable on the x-axis.)*
image.png

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
matrix = {
    'Mass': [0.45, 0.91, 1.36, 1.81, 2.27],
    'Fuel': [3.6, 6.7, 9.8, 11.2, 14.7]
}

cars = pd.DataFrame(matrix)

# Scatter plot
cars.plot(kind='scatter', x='Mass', y='Fuel', title='Scatter Plot (Fuel Used vs. Mass) ')
plt.show()

# Calculate the correlation coefficient
corr_type = cars['Mass'].corr(cars['Fuel'])
print("Correlation : ",corr_type)
# Determine the nature of the correlation
if corr_type > 0:
    corr_type = 'Positive'
else:
    correlation_type = 'Negative'


print(f'Correlation Type: {corr_type}')

```

##Since Correlation is close to 1 (0.993), we can say its a Positive Corelation

##If correlation coefficient is close to 1 or -1, it indicates a strong linear relationship

##Hence its a Linear Correlation

**Q3 The data below represents the number of chairs in each class of a government high school.**

**Create a box plot and swarm plot (add jitter) and find the number of data points that are outliers.**

**35, 54, 60, 65, 66, 67, 69, 70, 72, 73, 75, 76, 54, 25, 15, 60, 65, 66, 67, 69, 70, 72, 130, 73, 75, 76**

```python

chairs = [35, 54, 60, 65, 66, 67, 69, 70, 72, 73, 75, 76, 54, 25, 15, 60, 65, 66, 67, 69, 70, 72, 130, 73, 75, 76]

# Calculating quartiles and IQR
Q1 = np.percentile(chairs, 25)
Q3 = np.percentile(chairs, 75)
IQR = Q3 - Q1

#Calculating min and max values
min = Q1 - 1.5 * IQR
max = Q3 + 1.5 * IQR

outliers = []
for outlier in chairs:
  if outlier<min or outlier>max:
    outliers.append(outlier)


#box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=chairs)
plt.title("Box Plot Number of Chairs In Classes")
plt.xlabel("Number of Chairs")
plt.show()

#swarm plot with jitter
plt.figure(figsize=(8, 6))

# Adding jitter
jitter = 0.2
data_jittered = [val + np.random.uniform(-jitter, jitter) for val in chairs]

#swarm plot
sns.swarmplot(data=data_jittered, orient="v", color="blue", size=6)
plt.title("Swarm Plot of Number of Chairs in Classes")
plt.xlabel("Number of Chairs")
plt.show()

num_outliers = len(outliers)
print(f"Number of Outliers: {num_outliers}")
print("The outliers are : ",outliers)
```

##There are 4 outliers which are 35,25,15 and 130

**Q4 Generate random numbers from the following distribution and visualize the data using violin
plot.**

**(i) Standard-Normal distribution.**

**(ii) Log-Normal distribution.**

```python
# Generating random numbers from the standard-normal distribution
data_standard_normal = np.random.randn(1000)

# Generating random numbers from the log-normal distribution
data_log_normal = np.random.lognormal(mean=0, sigma=1, size=1000)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

#violin plots for the standard-normal and log-normal data
sns.violinplot(y=data_standard_normal, ax=axes[0], color="green", inner="quartile")
sns.violinplot(y=data_log_normal, ax=axes[1], color="orange", inner="quartile")

axes[0].set_title("Standard-Normal Distribution")
axes[1].set_title("Log-Normal Distribution")
axes[0].set_ylabel("Values")
axes[1].set_ylabel("Values")


plt.show()
```

**Q5 An Advertisement agency develops new ads for various clients (like Jewellery shops, Textile
shops).**

**The Agency wants to assess their performance, for which they want to know the number of ads they developed in each quarter for different shop category.**

**Help them to visualize data using radar/spider charts.**

image.png

```python
#creating a dataframe
shop = pd.DataFrame(
    {"Shop Category" : ["Textile","Jewellery","Cleaning Essentials","Cosmetics"],
    "Quarter 1" : [10,5,15,14],
    "Quarter 2" : [6,5,20,10],
    "Quarter 3" : [8,2,16,21],
    "Quarter 4" : [13,4,15,11]}
)
#print(shop)
categories = list(shop.columns[1:])

N = len(categories)

values = shop.iloc[:, 1:].values.tolist()


for val in values:
    val.append(val[0])

#creating angles for different quarters
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

#radar plot
plt.figure(figsize=(8, 6))
ax = plt.subplot(111, polar=True)

plt.xticks(angles[:-1], categories, color='grey', size=12)


ax.set_rlabel_position(0)
plt.yticks([5, 10, 15, 20], ["5","10","15","20"], color="grey", size=10)
plt.ylim(0, 25)

for i, row in enumerate(values):
    ax.plot(angles, row, linewidth=2, linestyle='solid', label=shop["Shop Category"][i])

ax.fill(angles, values[0], 'b', alpha=0.1)

plt.title("Radar/Spider Chart ")
.legend( bbox_to_anchor=(0.1, 0.1))
plt.show()
```

**Q6. An organization wants to calculate the % of time they spent on each process for their product
development.**

**Visualize the data using a funnel chart with the data given below.**


image.png

```python
import pandas as pd
import plotly.graph_objects as go

# Creating a DataFrame
data = {
    "Process Development Steps": ["Requirement Elicitation", "Requirement Analysis", "Software Development", "Debugging & Testing", "Others"],
    "Time Spent (in hours)": [50, 110, 250, 180, 70]
}

org = pd.DataFrame(data)

# Calculating the percentage of time spent for each step
total_time = org["Time Spent (in hours)"].sum()
org["Percentage"] = (org["Time Spent (in hours)"] / total_time) * 100

# Creating a funnel chart
fig = go.Figure(go.Funnel(
    y=org["Process Development Steps"],
    x=org["Percentage"],
    textinfo="percent total+percent previous"
))

fig.update_layout(
    title="Percentage of Time Spent on Product Development Processes  (Funnel Chart)",
    xaxis_title="Percentage",
    yaxis_title="Process Step",
    xaxis=dict(tickvals=[0, 25, 50, 75, 100]),
)

fig.show()
```

alter your purchasing pattern to match this trend, you want to be sure that the relationship is real.

**Q8 Given two arrays of numeric values, identify the suitable visualization mechanisms like Heat
Maps to draw a relationship between the given two sets. Also extend the same for the IRIS
dataset for possible attribute subset of your choice.**

Dataset: IRIS Dataset (https://www.kaggle.com/datasets/uciml/iris)

**1)Using Two Arrays**

```python
#IMPORT THE DATASET FROM THE ZIP FILE PROVIDED
# 1) USING RANDOM NUMERIC VALUE ARRAYS

#Randomised using random library
array1 = np.random.rand(100)
array2 = np.random.rand(100)

# Calculating the correlation matrix
corr_matrix = np.corrcoef(array1, array2)

# Create a heat map
plt.figure(figsize=(8, 6))
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar(label='Correlation')
plt.title('Heat Map of Correlation between Two Arrays')
plt.xticks(range(2), ['Array 1', 'Array 2'])
plt.yticks(range(2), ['Array 1', 'Array 2'])
plt.show()
```

As we increase the number of arrays the correlations between them changes and we can see distinct color variations


**2) Using IRIS Dataset**

```python
#reading the csv file and creating it into a dataframe
iris = pd.read_csv("Iris.csv")


corr_matrix = iris.corr()#finding the correlation matrix

# Plotting a heat map
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heat Map of Correlation in Iris Dataset')
plt.show()
```
