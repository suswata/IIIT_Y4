Dataset: IRIS Dataset (https://www.kaggle.com/datasets/uciml/iris)

It includes three iris species with 50 samples each as well as some properties about each flower. One
flower species is linearly separable from the other two, but the other two are not linearly separable from

The columns in this dataset are:

```python
#importing the iris dataset into a dataframe
df = pd.read_csv("Iris.csv")
print(df.head)
```

####BAR PLOT

```python
#species vs sepal length
sns.barplot(x="Species", y="SepalLengthCm", data=df)
plt.title("Bar Plot of Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.xticks(rotation=45)
plt.show()
#species vs sepal width
sns.barplot(x="Species", y="SepalWidthCm", data=df)
plt.title("Bar Plot of Sepal Width by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Width (cm)")
plt.xticks(rotation=45)
plt.show()
```

```python
#species vs petal length
sns.barplot(x="Species", y="PetalLengthCm", data=df, ci=None)
plt.title("Bar Plot of Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.xticks(rotation=45)
plt.show()
#species vs petal width
sns.barplot(x="Species", y="PetalWidthCm", data=df, ci=None)
plt.title("Bar Plot of Petal Width by Species")
plt.xlabel("Species")
plt.ylabel("Petal Width (cm)")
plt.xticks(rotation=45)
plt.show()
```

####Pie Chart


```python
#Pie Plot for Species
species_counts = df["Species"].value_counts()
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%')
plt.title("Pie Chart Species")
plt.show()

# Pie Plot of Petal Length by Species
species_petal_length = df.groupby("Species")["PetalLengthCm"].mean()
plt.pie(species_petal_length, labels=species_petal_length.index, autopct='%1.1f%%')
plt.title("Pie Chart of Petal Length by Species")
plt.show()
# Pie Plot of Petal Width by Species
species_petal_width = df.groupby("Species")["PetalWidthCm"].mean()
plt.pie(species_petal_width, labels=species_petal_width.index, autopct='%1.1f%%')
plt.title("Pie Chart of Petal Width by Species")
plt.show()
```

2. A Coach tracked the number of points that each of his 30 players on the team had in one game. The
points scored by each player is given below. Visualize the data using ordered stem-leaf plot and also
detect the outliers and shape of the distribution.


22, 21, 24, 19, 27, 28, 24, 25, 29, 28, 26, 31, 28, 27, 22, 39, 20, 10, 26, 24, 27, 28, 26, 28, 18, 32, 29,
25, 31, 27.

```python
import pandas as pd #importing libraries
import matplotlib.pyplot as plt


data = [22, 21, 24, 19, 27, 28, 24, 25, 29, 28, 26, 31, 28, 27, 22, 39, 20, 10, 26, 24, 27, 28, 26, 28, 18, 32, 29, 25, 31, 27]

series = pd.Series(data)

stem = series // 10
leaf = series % 10

stemleaf = pd.DataFrame({'Stem': stem, 'Leaf': leaf})

stemleaf.sort_values(['Stem', 'Leaf'], inplace=True)

stemleaf = stemleaf.groupby('Stem')['Leaf'].apply(list)

# Printing the stem-and-leaf plot
for s, l in stemleaf.items():
    print(f"{s}: {', '.join(map(str, l))}")

q1 = series.quantile(0.25)
q3 = series.quantile(0.75)
iqr = q3 - q1

min_end = q1 - 1.5 * iqr
max_end = q3 + 1.5 * iqr
outliers = series[(series < min_end) | (series > max_end)]
print("Outliers:", list(outliers))

plt.hist(data, bins=25, edgecolor='k', alpha=0.7)
plt.title("Histogram of Points Scored")
plt.xlabel("Points")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

####Curve Looks like The values obey Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.signal import find_peaks

data = [22, 21, 24, 19, 27, 28, 24, 25, 29, 28, 26, 31, 28, 27, 22, 39, 20, 10, 26, 24, 27, 28, 26, 28, 18, 32, 29, 25, 31, 27]

plt.hist(data, bins=10, edgecolor='k', alpha=0.7, density=True)
plt.title("Histogram of Points Scored")
plt.xlabel("Points")
plt.ylabel("Density")

#checking normal distribution
kde = norm.pdf(np.linspace(min(data), max(data), 1000), loc=np.mean(data), scale=np.std(data))

peaks, _ = find_peaks(kde)

plt.plot(np.linspace(min(data), max(data), 1000), kde, label='KDE', color='blue')

plt.vlines(np.linspace(min(data), max(data), 1000)[peaks], ymin=0, ymax=max(kde), color='red', linestyle='--', label='Peaks')
plt.legend()
plt.grid(True)
plt.show()
```
