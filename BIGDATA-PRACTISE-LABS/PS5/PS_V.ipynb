Q9. Test drive the use of Gini Index, Information Gain, Entropy and other measures that are supported in your
platform, performing the role of data selection.

Gini Index

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

data = pd.read_csv('adult.csv')
data_encoded = pd.get_dummies(data, columns=['workclass','education','marital-status','occupation','relationship','race','gender','native-country'])

X = data_encoded.drop(columns=['income'])  # Features
y = data_encoded['income']  # Target variable

clf = DecisionTreeClassifier()
clf.fit(X, y)

# Calculating the Gini index for each feature
gini_importance = clf.feature_importances_

for feature, importance in zip(X.columns, gini_importance):
    print(f'{feature}: {importance:.4f}')


```

Entropy

```python
tot_inst = len(data)
pos_inst = len(data[data['income'] == '>50K'])
neg_inst = tot_inst - pos_inst

p_positive = pos_inst / tot_inst
p_negative = neg_inst / tot_inst

entropy = -p_positive * np.log2(p_positive) - p_negative * np.log2(p_negative)

print(f'Entropy: {entropy:.4f}')
```

Information Gain

```python
information_gain = entropy #using the entropy
feature_name = 'education'
for feature_value in data[feature_name].unique():
    subset = data[data[feature_name] == feature_value]
    total_subset_instances = len(subset)
    positive_subset_instances = len(subset[subset['income'] == '>50K'])
    negative_subset_instances = total_subset_instances - positive_subset_instances
    p_positive_subset = positive_subset_instances / total_subset_instances
    p_negative_subset = negative_subset_instances / total_subset_instances
    entropy_subset = -p_positive_subset * np.log2(p_positive_subset) - p_negative_subset * np.log2(p_negative_subset)

    information_gain -= (total_subset_instances / tot_inst) * entropy_subset

# Print the Information Gain
print(f'Information Gain for feature "{feature_name}": {information_gain:.4f}')
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris

#using iris dataset
data = load_iris()
X, y = data.data, data.target

#Using Chi-squared (for classification problems)
chi2_selector = SelectKBest(chi2, k=2)  # Select the top 2 features
X_chi2_selected = chi2_selector.fit_transform(X, y)

#Using Mutual Information (for classification problems)
mi_selector = SelectKBest(mutual_info_classif, k=2)  # Select the top 2 features
X_mi_selected = mi_selector.fit_transform(X, y)

#Using Tree-based feature selection (Random Forest)
rf_selector = SelectFromModel(RandomForestClassifier(n_estimators=100))
X_rf_selected = rf_selector.fit_transform(X, y)

print("Chi-squared selected features:")
print(X_chi2_selected)
print("------------------------------")

print("\nMutual Information selected features:")
print(X_mi_selected)
print("------------------------------")

print("\nRandom Forest selected features:")
print(X_rf_selected)

```
